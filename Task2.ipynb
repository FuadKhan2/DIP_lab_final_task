```python
# Task 2
from skimage import io, color
import numpy as np
import matplotlib.pyplot as plt

# Load the RGB image
image_path = r'C:\Users\user\Downloads\DIP_Lab_task_02.jpg'  # Replace with your image path
image_rgb = io.imread(image_path)  # Load the image in RGB format

# Convert to grayscale
image_gray = color.rgb2gray(image_rgb)  # Convert to grayscale (values normalized to [0, 1])
image_gray = (image_gray * 255).astype(np.uint8)  # Scale to 8-bit integer values (0255)

# Extract a sub-region (5x5) from both RGB and grayscale images
# Select a region starting at (50, 50) for example
start_x, start_y = 50, 50
sub_region_rgb = image_rgb[start_y:start_y + 5, start_x:start_x + 5]
sub_region_gray = image_gray[start_y:start_y + 5, start_x:start_x + 5]

# Display the original RGB and grayscale images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original RGB Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray', vmin=0, vmax=255)
plt.title("Grayscale Image")
plt.axis('off')

plt.show()

# Display the sub-region values
print("Sub-region (5x5) from RGB Image:")
print(sub_region_rgb)

print("\nSub-region (5x5) from Grayscale Image:")
print(sub_region_gray)

# Visualize the 5x5 grayscale sub-region
plt.figure(figsize=(4, 4))
plt.imshow(sub_region_gray, cmap='gray', vmin=0, vmax=255)
plt.title("5x5 Sub-region of Grayscale Image")
plt.axis('on')
plt.show()
```
